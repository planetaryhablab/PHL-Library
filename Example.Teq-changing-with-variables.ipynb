Using the program, it's also possible to graphicate the temperature in terms of the variables.

```python
import phl
import matplotlib.pyplot as plt
import numpy as np
```

```python
e = np.linspace(0, 1, 100)
K = [phl.teq(1.0,e,0.3,1.0)for e in e]
```

```python
fig = plt.figure()
axes = fig.add_subplot(111)
axes.plot(e,K)
plt.show()
```

The graph above shows how the equilibrium temperature changes in terms of eccentricity in the orbit of the planet.

```python
A = np.linspace(0, 1, 100)
K = [phl.teq(1.0,0.0,A,1.0)for A in A]
```

```python
fig = plt.figure()
axes = fig.add_subplot(111)
axes.plot(A,K)
plt.show()
```

This other graph, shows the change of the equilibrium temperature in terms of the albedo of the planet.
