# In this file, the functions and uses of the program are explained via examples.

```python
import phl
import matplotlib.pyplot as plt
import numpy as np
```

First, we are using the Teq function

```python
a  = np.linspace(0.1, 1, 100)
K  = [phl.teq(a,0.0,0.3,1.0)for a in a]

e  = np.linspace(0, 1, 100)
K2 = [phl.teq(1.0,e,0.3,1.0)for e in e]

A  = np.linspace(0, 1, 100)
K3 = [phl.teq(1.0,0.0,A,1.0)for A in A]

L  = np.linspace(0, 1, 100)
K4 = [phl.teq(1.0,0.0,0.3,L)for L in L]
```

In this example, we compared how the equilibrium temperature would change in terms of all the variables using graphs of the functions and Earth as a reference

```python
fig  = plt.figure()
axes = fig.add_subplot(111)
axes.plot(a,K)
axes.plot(e,K2)
axes.plot(A,K3)
axes.plot(L,K4)
```

In this other example we compare how the equilibrium temperature would change in terms of Luminosity only with all planets of the solar system

```python
A = np.linspace(0, 1, 100)
Mercury = [phl.teq(0.4,0.2,A,1.0)for A in A]
Venus   = [phl.teq(0.7,0.0,A,1.0)for A in A]
Earth   = [phl.teq(1.0,0.0,A,1.0)for A in A]
Mars    = [phl.teq(1.5,0.1,A,1.0)for A in A]
Jupiter = [phl.teq(5.2,0.0,A,1.0)for A in A]
Saturn  = [phl.teq(9.5,0.0,A,1.0)for A in A]
Uranus  = [phl.teq(19,0.0,A,1.0)for A in A]
Neptune = [phl.teq(30,0.1,A,1.0)for A in A]

fig  = plt.figure()
axes = fig.add_subplot(111)
axes.plot(A,Mercury)
axes.plot(A,Venus)
axes.plot(A,Earth)
axes.plot(A,Mars)
axes.plot(A,Jupiter)
axes.plot(A,Saturn)
axes.plot(A,Uranus)
axes.plot(A,Neptune)
plt.show()
```

### Using the program, it's also possible to detect the orbital effective thermal distance in AU

In this example we detect and graphicate the REFF of the planets of the solar system using the semi-major axis and eccentrity

```python
plt.plot([1, 2, 3, 4, 5, 6, 7, 8], [phl.reff(0.39,0.21),phl.reff(0.72,0.006),phl.reff(1.0,0.00),phl.reff(1.5,0.09),phl.reff(5.2,0.04),phl.reff(9.5,0.05),phl.reff(19.19,0.05),phl.reff(30.06,0.06)], 'ro')
plt.axis([0, 9, 0, 35])
plt.show()
```

### Using the program, it's also possible to detect orbital average stellar flux in AU

In this example, we are detecting the FLUX of planet Earth if Luminosity was changed.

```python
L = np.linspace(0, 1, 100)
Earth   = [phl.flux(1.0,0.0,L)for L in L]
fig  = plt.figure()
axes = fig.add_subplot(111)
axes.plot(L,Earth)

plt.show()
```

In this other example, we are detecting and graphicating the change FLUX of the planets in the solar system and how the change in Eccentricity would change the results.

```python
e = np.linspace(0, 0.9, 100)
Mercury = [phl.flux(0.4,e,1.0)for e in e]
Venus   = [phl.flux(0.7,e,1.0)for e in e]
Earth   = [phl.flux(1.0,e,1.0)for e in e]
Mars    = [phl.flux(1.5,e,1.0)for e in e]
Jupiter = [phl.flux(5.2,e,1.0)for e in e]
Saturn  = [phl.flux(9.5,e,1.0)for e in e]
Uranus  = [phl.flux(19,e,1.0)for e in e]
Neptune = [phl.flux(30,e,1.0)for e in e]
```

```python
fig  = plt.figure()
axes = fig.add_subplot(111)
axes.plot(e,Mercury)
axes.plot(e,Venus)
axes.plot(e,Earth)
axes.plot(e,Mars)
axes.plot(e,Jupiter)
axes.plot(e,Saturn)
axes.plot(e,Uranus)
axes.plot(e,Neptune)
plt.show()
```
