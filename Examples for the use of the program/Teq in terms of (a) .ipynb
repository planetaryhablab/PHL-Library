Using the program, it's also possible to graphicate the temperature in terms of the variables.

```python
import phl
import matplotlib.pyplot as plt
import numpy as np
```

(For the sake of simplicity, all of the variable's values are approximated to 2 digits)

```python
a = np.linspace(0.1, 1, 100)
Mercury = [phl.teq(a,0.2,0.1,1.0)for a in a]
Venus   = [phl.teq(a,0.0,0.9,1.0)for a in a]
Earth   = [phl.teq(a,0.0,0.3,1.0)for a in a]
Mars    = [phl.teq(a,0.1,0.2,1.0)for a in a]
Jupiter = [phl.teq(a,0.0,0.5,1.0)for a in a]
Saturn  = [phl.teq(a,0.0,0.4,1.0)for a in a]
Uranus  = [phl.teq(a,0.0,0.5,1.0)for a in a]
Neptune = [phl.teq(a,0.1,0.4,1.0)for a in a]
```

In this example, we used our solar system's planets as references and graphicated how the equilibrium temperature would change in terms of the semi-mayor axis (Using astronomical units).

```python
fig = plt.figure()
axes = fig.add_subplot(111)
axes.plot(a,Mercury)
axes.plot(a,Venus)
axes.plot(a,Earth)
axes.plot(a,Mars)
axes.plot(a,Jupiter)
axes.plot(a,Saturn)
axes.plot(a,Uranus)
axes.plot(a,Neptune)
plt.show()
```
