Using the program, it's also possible to graphicate the temperature in terms of the variables.

```python
import phl
import matplotlib.pyplot as plt
import numpy as np
```

(For the sake of simplicity, all of the variable's values are approximated to 2 digits)

```python
e = np.linspace(0, 1, 100)
Mercury = [phl.teq(0.4,e,0.1,1.0)for e in e]
Venus   = [phl.teq(0.7,e,0.9,1.0)for e in e]
Earth   = [phl.teq(1.0,e,0.3,1.0)for e in e]
Mars    = [phl.teq(1.5,e,0.2,1.0)for e in e]
Jupiter = [phl.teq(5.2,e,0.5,1.0)for e in e]
Saturn  = [phl.teq(9.5,e,0.4,1.0)for e in e]
Uranus  = [phl.teq(19,e,0.5,1.0)for e in e]
Neptune = [phl.teq(30,e,0.4,1.0)for e in e]
```

In this example, we used our solar system's planets as references and graphicated how the equilibrium temperature would change in terms of the orbital eccentricity.

```python
fig  = plt.figure()
axes = fig.add_subplot(111)
axes.plot(e,Mercury)
axes.plot(e,Venus)
axes.plot(e,Earth)
axes.plot(e,Mars)
axes.plot(e,Jupiter)
axes.plot(e,Saturn)
axes.plot(e,Uranus)
axes.plot(e,Neptune)
plt.show()
```
