Using the program, it's also possible to graphicate the temperature in terms of the variables.

```python
import phl
import matplotlib.pyplot as plt
import numpy as np
```

(For the sake of simplicity, all of the variable's values are approximated to 2 digits)

```python
L = np.linspace(0, 1, 100)
Mercury = [phl.teq(0.4,0.2,0.1,L)for L in L]
Venus   = [phl.teq(0.7,0.0,0.9,L)for L in L]
Earth   = [phl.teq(1.0,0.0,0.3,L)for L in L]
Mars    = [phl.teq(1.5,0.1,0.2,L)for L in L]
Jupiter = [phl.teq(5.2,0.0,0.5,L)for L in L]
Saturn  = [phl.teq(9.5,0.0,0.4,L)for L in L]
Uranus  = [phl.teq(19,0.0,0.5,L)for L in L]
Neptune = [phl.teq(30,0.1,0.4,L)for L in L]
```

In this example, we used our solar system's planets as references and graphicated how the equilibrium temperature would change in terms of the stellar luminosity (using solar units).

```python
fig  = plt.figure()
axes = fig.add_subplot(111)
axes.plot(L,Mercury)
axes.plot(L,Venus)
axes.plot(L,Earth)
axes.plot(L,Mars)
axes.plot(L,Jupiter)
axes.plot(L,Saturn)
axes.plot(L,Uranus)
axes.plot(L,Neptune)
plt.show()
```
