Using the program, it's also possible to graphicate the temperature in terms of the variables.

```python
import phl
import matplotlib.pyplot as plt
import numpy as np
```

(For the sake of simplicity, all of the variable's values are approximated to 2 digits)

```python
A = np.linspace(0, 1, 100)
Mercury = [phl.teq(0.4,0.2,A,1.0)for A in A]
Venus   = [phl.teq(0.7,0.0,A,1.0)for A in A]
Earth   = [phl.teq(1.0,0.0,A,1.0)for A in A]
Mars    = [phl.teq(1.5,0.1,A,1.0)for A in A]
Jupiter = [phl.teq(5.2,0.0,A,1.0)for A in A]
Saturn  = [phl.teq(9.5,0.0,A,1.0)for A in A]
Uranus  = [phl.teq(19,0.0,A,1.0)for A in A]
Neptune = [phl.teq(30,0.1,A,1.0)for A in A]
```

In this example, we used our solar system's planets as references and graphicated how the equilibrium temperature would change in terms of the planetary bond albedo.

```python
fig  = plt.figure()
axes = fig.add_subplot(111)
axes.plot(A,Mercury)
axes.plot(A,Venus)
axes.plot(A,Earth)
axes.plot(A,Mars)
axes.plot(A,Jupiter)
axes.plot(A,Saturn)
axes.plot(A,Uranus)
axes.plot(A,Neptune)
plt.show()
```
